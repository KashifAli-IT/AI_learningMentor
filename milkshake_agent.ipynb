#  🧍 Human Example: Making a Milkshake

Let’s say we give a human a goal:

👉 “Make a milkshake”

### Step 1: Think 🤔
- Do I have bananas?
- Do I have milk?
- Do I have sugar?

### Step 2: Act 🛠️
- Go to kitchen
- If something missing → go buy it

### Step 3: Observe 👀
- Are ingredients good?
- Is something missing?

### Step 4: Repeat 🔁
- Fix issues
- Try again

### Step 5: Improve ✅
- Taste it
- Adjust sugar or milk

---

## 🔁 This is a LOOP

**Think → Act → Observe → Repeat**

👉 This is exactly how AI Agents work

</details>

## 💡 Key Idea

So instead of just answering…

👉 The agent will:
- Decide what to do
- Use tools
- Fix mistakes
- Try again

</details>

```python
from smolagents.models import ChatMessage

class GroqModel:
    def __init__(self, client, model="llama-3.3-70b-versatile"):
        self.client = client
        self.model = model

    def generate(self, messages, stop_sequences=None):
        formatted_messages = [
            {
                "role": m.role if m.role in ["system", "user", "assistant"] else "assistant",
                "content": m.content
            }
            for m in messages
        ]

        response = self.client.chat.completions.create(
            model=self.model,
            messages=formatted_messages,
            temperature=0
        )

        # ✅ RETURN ChatMessage (what CodeAgent expects)
        return ChatMessage(
            role="assistant",
            content=response.choices[0].message.content
        )
```

# 🛠️ Tools = Agent’s Hands

Humans use:
- Hands
- Eyes
- Actions

AI Agents use:
- Functions (tools)

---

## 🧪 In This Demo

We give the agent these tools:

- Check ingredients → (Look in kitchen)
- Buy ingredient → (Go to market)
- Make milkshake → (Final action)

---

## 💡 Key Idea

> Tools turn AI from “thinking” → into “doing”
</details>

```python
# Kitchen state (start with missing banana)
kitchen = {
    "banana": False,
    "milk": False,
    "sugar": True
}

@tool
def check_ingredients() -> dict:
    """
    Check available ingredients in the kitchen.

    Returns:
        dict: A dictionary of ingredients and their availability.
    """
    print("👀 Agent is checking ingredients from fridge/kitchen...")
    return kitchen


@tool
def buy_ingredient(item: str) -> str:
    """
    Buy a missing ingredient.

    Args:
        item (str): Name of the ingredient to purchase.

    Returns:
        str: Confirmation message after buying the ingredient.
    """
    print(f"🧠 Agent decided to buy {item}")
    print(f"🛒 Buying {item}...")
    kitchen[item] = True
    return f"{item} purchased"


@tool
def make_milkshake() -> str:
    """
    Make milkshake if all ingredients are available.

    Returns:
        str: Result of milkshake preparation.
    """
    print("🥤 Attempting to make milkshake...")
    print("🥤 Agent is making milkshake...")

    if all(kitchen.values()):
        return "Milkshake is ready! 🎉"
    else:
        return "Missing ingredients, cannot make milkshake"
```

# 🧠 The Brain (LLM)

The brain of the agent is an AI model (like LLaMA).

Its job:
- Understand the goal
- Decide next step
- Choose which tool to use

---

## 💡 Important

The model is NOT pre-programmed with steps.

👉 It decides dynamically:
- What to do first
- What to do next
- When to stop
</details>

```python
llm = GroqModel(client)

agent = CodeAgent(
    tools=[check_ingredients, buy_ingredient, make_milkshake],
    model=llm,
    verbosity_level=2,
    max_steps=5,
    instructions="""
You are Ayaan, a helpful and smart assistant.

You help in the kitchen by:
1. Checking ingredients
2. Buying missing items
3. Completing the task

You think step by step and use tools when needed.
"""
)
```
